```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModel
from rdkit import Chem
import pickle
from tqdm import tqdm

# ========== æ–‡ä»¶è·¯å¾„ ==========
file_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\smiles.xlsx"
embedding_output_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\chemberta_embeddings.csv"
graph_output_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\molecular_graphs.pkl"

# ========== æ•°æ®è¯»å– ==========
df = pd.read_excel(file_path)
name_col = "material_name"
smiles_col = "SMILES"

# ========== åŠ è½½ ChemBERTa æ¨¡å‹ ==========
tokenizer = AutoTokenizer.from_pretrained("DeepChem/ChemBERTa-77M-MTR")
model = AutoModel.from_pretrained("DeepChem/ChemBERTa-77M-MTR")
device = torch.device("cpu")
model.to(device)
model.eval()

# ========== å­˜å‚¨å˜é‡ ==========
embedding_list = []
valid_names = []
graph_dict = {}

# ========== å¤„ç†æ¯ä¸ª SMILES ==========
for _, row in tqdm(df.iterrows(), total=len(df)):
    name = row[name_col]
    smiles = row[smiles_col]

    if pd.isna(smiles) or smiles.strip() == "":
        continue

    try:
        # ====== ChemBERTa embedding ======
        inputs = tokenizer(smiles, return_tensors="pt", truncation=True, padding=True).to(device)
        with torch.no_grad():
            outputs = model(**inputs)
            embedding = outputs.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()

        embedding_list.append(embedding)
        valid_names.append(name)

        # ====== åˆ†å­å›¾ç»“æ„æå– (RDKit) ======
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            edge_index = []
            for bond in mol.GetBonds():
                a1 = bond.GetBeginAtomIdx()
                a2 = bond.GetEndAtomIdx()
                edge_index.append((a1, a2))
                edge_index.append((a2, a1))  # åŒå‘è¾¹
            graph_dict[name] = {
                "num_atoms": mol.GetNumAtoms(),
                "edge_index": edge_index
            }

    except Exception as e:
        print(f"[ERROR] {name} skipped: {e}")

# ========== ä¿å­˜ embedding ==========
latent_cols = [f"latent_{i}" for i in range(len(embedding_list[0]))]
embedding_df = pd.DataFrame(embedding_list, columns=latent_cols)
embedding_df.insert(0, "material_name", valid_names)
embedding_df.to_csv(embedding_output_path, index=False)
print(f"âœ… Embedding saved to: {embedding_output_path}")

# ========== ä¿å­˜åˆ†å­å›¾ ==========
with open(graph_output_path, "wb") as f:
    pickle.dump(graph_dict, f)
print(f"âœ… Molecular graphs saved to: {graph_output_path}")

```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm

# === è·¯å¾„ ===
main_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\2_regression_original.xlsx"
embedding_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\chemberta_embeddings.csv"
output_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\regression_final_all_with_weighted_embedding.xlsx"

# === è¯»å–æ•°æ® ===
df_main = pd.read_excel(main_path)
df_embed = pd.read_csv(embedding_path)

# === æ„å»º embedding å­—å…¸ï¼Œå¹¶ç¡®ä¿ä¸º float ç±»å‹ ===
latent_cols = [f"latent_{i}" for i in range(384)]
embedding_dict = {
    row["material_name"]: row[latent_cols].values.astype(float)
    for _, row in df_embed.iterrows()
}

# === åˆå§‹åŒ–è¾“å‡º embedding åˆ—è¡¨ ===
weighted_embeddings = []

# === éå†æ¯ä¸€æ¡é…æ–¹ ===
for _, row in tqdm(df_main.iterrows(), total=len(df_main)):
    weighted_sum = np.zeros(384, dtype=float)
    total_weight = 0.0

    for col in df_main.columns:
        if col in embedding_dict:
            amount = row[col]
            if pd.notna(amount) and amount > 0:
                embedding = embedding_dict[col]  # ç°åœ¨å·²ç»ç¡®ä¿æ˜¯ float ç±»å‹äº†
                weighted_sum += embedding * amount
                total_weight += amount

    if total_weight > 0:
        weighted_avg = weighted_sum / total_weight
    else:
        weighted_avg = np.zeros(384, dtype=float)

    weighted_embeddings.append(weighted_avg)

# === æ„å»º embedding DataFrame å¹¶æ‹¼æ¥å›ä¸»è¡¨ ===
df_embedding = pd.DataFrame(weighted_embeddings, columns=latent_cols)
df_final = pd.concat([df_main.reset_index(drop=True), df_embedding], axis=1)

# === ä¿å­˜ç»“æœ ===
df_final.to_excel(output_path, index=False)
print(f"âœ… åŠ æƒ embedding å·²ä¿å­˜åˆ°ï¼š{output_path}")

```

```python
from rdkit.Chem import AllChem
from rdkit.Chem.rdFingerprintGenerator import GetMorganGenerator

# === è·¯å¾„ ===
embedding_main_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\regression_final_all_with_weighted_embedding.xlsx"
smiles_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\smiles.xlsx"
output_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\regression_with_chemberta_rdkit.xlsx"

# === å‚æ•° ===
fp_size = 128
fp_radius = 2

# === è¯»å–æ•°æ®è¡¨ ===
df_main = pd.read_excel(embedding_main_path)
df_smiles = pd.read_excel(smiles_path)

# === æ„å»º SMILES â†’ RDKit fingerprint å­—å…¸ ===
fp_dict = {}
generator = GetMorganGenerator(radius=fp_radius, fpSize=fp_size) 

for _, row in df_smiles.iterrows():
    name = row["material_name"]
    smiles = row["SMILES"]
    if pd.notna(smiles) and isinstance(smiles, str) and smiles.strip() != "":
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            fp = generator.GetFingerprint(mol).ToList() 
            arr = np.array(fp).astype(float)            
            fp_dict[name] = arr

# === æ„é€ å„é…æ–¹çš„åŠ æƒ fingerprint ===
fp_list = []

for _, row in tqdm(df_main.iterrows(), total=len(df_main)):
    weighted_sum = np.zeros(fp_size, dtype=float)
    total_weight = 0.0

    for col in df_main.columns:
        if col in fp_dict:
            amount = row[col]
            if pd.notna(amount) and amount > 0:
                fp = fp_dict[col]
                weighted_sum += fp * amount
                total_weight += amount

    if total_weight > 0:
        avg_fp = weighted_sum / total_weight
    else:
        avg_fp = np.zeros(fp_size, dtype=float)

    fp_list.append(avg_fp)

# === è½¬æˆ DataFrame å¹¶æ‹¼æ¥ä¸»è¡¨ ===
fp_cols = [f"rdkit_{i}" for i in range(fp_size)]
df_fp = pd.DataFrame(fp_list, columns=fp_cols)

df_final = pd.concat([df_main.reset_index(drop=True), df_fp], axis=1)

# === ä¿å­˜ ===
df_final.to_excel(output_path, index=False)
print(f"âœ… æœ€ç»ˆæ•°æ®è¡¨ï¼ˆå« ChemBERTa + RDKit ç­‰ç‰¹å¾ï¼‰å·²ä¿å­˜åˆ°: {output_path}")
```

```python
import pandas as pd
import numpy as np
from rdkit import Chem
from mordred import Calculator, descriptors
from tqdm import tqdm

# === è·¯å¾„ ===
main_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\regression_final_all_with_weighted_embedding.xlsx"
smiles_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\smiles.xlsx"
output_path = r"D:\ML-3DPrinting-Project\data\7.9_datasets\regression_with_mordred.xlsx"

# === è¯»å–ä¸»æ•°æ®é›†å’Œ SMILES æ•°æ® ===
df_main = pd.read_excel(main_path)
df_smiles = pd.read_excel(smiles_path)

# === åˆå§‹åŒ– Mordred calculator ===
calc = Calculator(descriptors, ignore_3D=True)

# === æ„å»ºæè¿°ç¬¦å­—å…¸ï¼ˆname â†’ Mordredç‰¹å¾ï¼‰ ===
desc_dict = {}
valid_features = None  # ç”¨äºåé¢ä¸€è‡´åŒ–åˆ—

print("ğŸ§ª æ­£åœ¨æå– Mordred ç‰¹å¾...")
for _, row in tqdm(df_smiles.iterrows(), total=len(df_smiles)):
    name = row["material_name"]
    smiles = row["SMILES"]
    if pd.isna(smiles) or not isinstance(smiles, str) or smiles.strip() == "":
        continue

    mol = Chem.MolFromSmiles(smiles)
    if mol:
        try:
            desc = calc(mol)
            desc_values = pd.Series(desc.asdict()).replace([np.inf, -np.inf], np.nan).astype(float)

            # åˆå§‹åŒ–åˆ—å
            if valid_features is None:
                valid_features = desc_values.dropna().index.tolist()

            # ä¿è¯æ‰€æœ‰æˆåˆ†æœ‰ç›¸åŒç»´åº¦ï¼ˆç¼ºå¤±å€¼å¡«0ï¼‰
            vector = desc_values[valid_features].fillna(0).values.astype(float)
            desc_dict[name] = vector
        except Exception as e:
            print(f"[âŒ] {name} è·³è¿‡: {e}")

# === åŠ æƒèåˆæˆé…æ–¹çº§ç‰¹å¾ ===
final_features = []

for _, row in tqdm(df_main.iterrows(), total=len(df_main)):
    weighted_sum = np.zeros(len(valid_features), dtype=float)
    total_weight = 0.0

    for col in df_main.columns:
        if col in desc_dict:
            amount = row[col]
            if pd.notna(amount) and amount > 0:
                weighted_sum += desc_dict[col] * amount
                total_weight += amount

    if total_weight > 0:
        avg_vector = weighted_sum / total_weight
    else:
        avg_vector = np.zeros(len(valid_features), dtype=float)

    final_features.append(avg_vector)

# === æ„å»ºç‰¹å¾è¡¨å¹¶æ‹¼æ¥å›ä¸»æ•°æ®é›† ===
desc_feature_names = [f"mordred_{name}" for name in valid_features]
df_desc = pd.DataFrame(final_features, columns=desc_feature_names)

df_final = pd.concat([df_main.reset_index(drop=True), df_desc], axis=1)

# === ä¿å­˜è¾“å‡º ===
df_final.to_excel(output_path, index=False)
print(f"âœ… èåˆåæ•°æ®é›†å·²ä¿å­˜åˆ°ï¼š{output_path}")
```

```python
import pandas as pd
from rdkit import Chem
from mordred import Calculator, descriptors
import numpy as np

# === 1. è¯»å…¥ SMILES æ–‡ä»¶ ===
df = pd.read_excel("D:/ML-3DPrinting-Project/data/7.9_datasets/smiles.xlsx")

# === 2. åˆå§‹åŒ– Mordred ç‰¹å¾æå–å™¨ ===
calc = Calculator(descriptors, ignore_3D=True)

# === 3. æ£€æŸ¥åˆ—è¡¨ ===
errors = []

for i, row in df.iterrows():
    name = row["material_name"]
    smiles = row["SMILES"]

    if pd.isna(smiles) or smiles.strip() == "":
        continue

    # 1ï¼‰SMILES è½¬åˆ†å­
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        errors.append((name, smiles, "MolFromSmiles failed"))
        continue

    # 2ï¼‰Kekulize å¤±è´¥æ£€æŸ¥
    try:
        Chem.Kekulize(mol, clearAromaticFlags=True)
    except:
        errors.append((name, smiles, "Kekulize failed"))

    # 3ï¼‰Mordred ç‰¹å¾æ­£å¸¸æ€§æ£€æŸ¥ï¼ˆä¿®å¤ç‰ˆï¼‰
    try:
        desc = calc(mol)
        desc_series = pd.Series(desc.asdict())

        # å°†ä¸èƒ½è½¬æ¢ä¸ºæ•°å­—çš„å­—æ®µè®¾ä¸º NaN
        desc_clean = desc_series.apply(lambda x: float(x) if isinstance(x, (int, float)) else np.nan)

        # åˆ¤æ–­æ˜¯å¦å­˜åœ¨å¼‚å¸¸å¤§å€¼
        if (desc_clean > 1e10).any():
            errors.append((name, smiles, "Mordred overflow"))

    except Exception as e:
        errors.append((name, smiles, f"Mordred error: {str(e)}"))

# === 4. å¯¼å‡ºæŠ¥é”™ç»“æœ ===
error_df = pd.DataFrame(errors, columns=["material_name", "SMILES", "error"])
error_df.to_excel("D:/ML-3DPrinting-Project/data/invalid_molecules_report.xlsx", index=False)
print(f"âš ï¸ å…±æœ‰ {len(error_df)} ä¸ªå¼‚å¸¸åˆ†å­ï¼Œç»“æœå·²å¯¼å‡ºã€‚")
```
